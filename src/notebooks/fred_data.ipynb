```r
library(quantmod)
library(data.table)
```

```r
fred.sources <- c('FEDFUNDS','GDP','UNRATE','PAYEMS','ICSA','CPIAUCSL','USREC')
```

```r
fred.list <- list()
for (fred.series in fred.sources) {
    print(fred.series)
    fred.list[[fred.series]] <- getSymbols(fred.series,src='FRED',return.class='xts', auto.assign = FALSE)
}
```

```r
fred.combined <- do.call(cbind, fred.list)
```

```r
#carry forward last observation
fred.combined <- na.locf(fred.combined["1970/"])
```

```r
write.csv(as.data.frame(fred.combined), "~/school/gt/mgt6203/project/Team-14/data/fred.csv")
```
